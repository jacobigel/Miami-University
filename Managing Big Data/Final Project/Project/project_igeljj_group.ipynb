```python
from statsmodels.stats.proportion import proportions_ztest

sample_success = 128.12
sample_size = 397

# our Ho is 51% * .66
null_hypothesis = (0.51 *.66)

# check our sample against Ho for Ha > Ho
# for Ha < Ho use alternative='smaller'
# for Ha != Ho use alternative='two-sided'
stat, p_value = proportions_ztest(count=sample_success, nobs=sample_size, value=null_hypothesis, alternative='two-sided')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

sample_success = 0.01
sample_size = 58

# our Ho is 48% * .66
null_hypothesis = (0.48 * .66)

# check our sample against Ho for Ha > Ho
# for Ha < Ho use alternative='smaller'
# for Ha != Ho use alternative='two-sided'
stat, p_value = proportions_ztest(count=sample_success, nobs=sample_size, value=null_hypothesis, alternative='two-sided')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))
```

The negative scores are re-evaluted to be a small positive number so that the python function would function properly. For the queries executed with times after the election was called, each z test is comparing each of the total scores to the expected value (the percentage of votes that the candidate recieved) with the number of nonzero scores.
