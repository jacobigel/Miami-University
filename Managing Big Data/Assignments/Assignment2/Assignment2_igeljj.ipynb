### Question 1

The above function receives a URL as an argument and returns a data frame called
result containing relevant data in five columns, namely Home_Team, Away_Team,
Winner_Team, Prediction_Home_Team, and Prediction_Away_Team. Your job is
now to fill the gap (i.e., the “…”). That is, use regular expressions to retrieve the
data of interest from the HTML code inside the variable html_code and to build the
result data frame.

```python
def scrap_football_data (url):
 import requests
 import re
 import pandas as pd
 r = requests.get(url)
 html_code = r.text

# Establishing all of the variables
 home_team = []
 away_team = []
 winner_team =[]
 prediction_home_team = []
 prediction_away_team = []

 matches_open_team = re.finditer("_BLANK\">(.+?)</A>", html_code) 

# Getting all teams
 all_teams = [m.group(1) for m in matches_open_team] 
 all_teams = [re.sub("<B>","", all_teams[i]) for i in range(len(all_teams))]
 all_teams = [re.sub("</B>","", all_teams[i]) for i in range(len(all_teams))]
# all_teams = [re.sub("<B>|</B>", '', all_teams[i] for i in range(len(allteams))]

 # Home Team
 home_team = all_teams[0::2]

 #Away Team
 away_team = all_teams[1::2]

# Winners
 b_tag = re.finditer("_BLANK\"><B>(.+?)</B></A>", html_code) 
 winner = [m.group(1) for m in b_tag]
 winner.pop(len(winner) - 1)
 winner_team = winner


# Getting all of the picks
 pick = re.finditer("_BLANK\">(.+?)<NOBR>(.+?)</NOBR>", html_code)
 all_pick = [m.group(2) for m in pick]
 all_pick.pop(len(all_pick) - 1)
 all_pick = [re.sub("<B>","", all_pick[i]) for i in range(len(all_pick))]
 all_pick = [re.sub("</B>","", all_pick[i]) for i in range(len(all_pick))]
 all_pick = [re.sub("%","", all_pick[i]) for i in range(len(all_pick))]

 # Home Team Predictions
 prediction_home_team = all_pick[0::2]

 #Away Team Predictions
 prediction_away_team = all_pick[1::2]
 

 # Creating the DataFrame
 result = pd.DataFrame(list(zip(home_team,away_team,winner_team,prediction_home_team,prediction_away_team)), columns=['Home_Team','Away_Team', 'Winner_Team','Prediction_Home_Team','Prediction_Away_Team'])
 result

 
 return result

# Test to make sure our function works on one link
scrap_football_data("https://probabilityfootball.com/picks.html?1487349677&username=AVERAGES&weeknum=5")
```
