### Question 1

```python
import pandas as pd
import re


file = open('access.log.txt', 'r')
access_logs = file.readlines()

# To check if the access_logs is working
# words = access_logs[0].split(" ")
# print(words)

```

***
###  <u>Q1-a)</u> IP

```python
IP = [access_logs[i].split(" ")[0] for i in range(len(access_logs))]
IP
```

***
###  <u>Q1-b)</u> Date

```python
dates = []

all_pos_first_bracket = [re.search("\[", log).start(0) for log in access_logs]
# all_pos_first_bracket

all_pos_second_bracket = [re.search("\]", log).start(0) for log in access_logs]
# all_pos_second_bracket

dates = [access_logs[i][all_pos_first_bracket[i] +1 :all_pos_second_bracket[i]:1] for i in range(len(access_logs))] # this will look at every index position
dates
```

***
###  <u>Q1-c)</u> Request

```python
request = []

request = [access_logs[i].split(" ")[5].replace("\"", "") for i in range(len(access_logs))]

request
```

***
### <u>Q1-d)</u> Page 

```python
page = []

page = [access_logs[i].split(" ")[6] for i in range(len(access_logs))]

page
```

***
###  <u>Q1-e)</u> HTTP_Version

```python
version = [access_logs[i].split(" ")[7].replace("\"","") for i in range(len(access_logs))]
version
```

***
### <u>Q1-f)</u> First_Code

```python
first_code = [access_logs[i].split(" ")[8] for i in range(len(access_logs))]
first_code
```

***
###  <u>Q1-g)</u> Second_Code

```python
second_code = [access_logs[i].split(" ")[9] for i in range(len(access_logs))]
second_code
```

***
### <u>Q1-h)</u> User_Agent

```python
user_agent = [access_logs[i].split(" ")[11]for i in range(len(access_logs))]
# user_agent 

all_pos_moz = [re.search("Mozilla", log).start(0) for log in access_logs]
# all_pos_moz

user_agent = [access_logs[i][all_pos_moz[i]::1].replace("\"\n", "") for i in range(len(access_logs))] # this will look at every index position

user_agent
```

***
### <u>Q1-i)</u> Creating a Dataframe

```python
data = pd.DataFrame(list(zip(IP,dates,request,page,version,first_code,second_code,user_agent)), columns=['IP','Date', 'Request','Page','HTTP_Version','First_Code','Second_Code', 'User_Agent'])
data
```

***
###  <u>Q1-j)</u> Exporting Dataframe to a CSV file

```python
data.to_csv("data.csv", index = False)
```
